```python
with open('data.txt',encoding='utf-8') as f:
    data = f.readlines()
```

```python
import jieba
from collections import Counter
```

```python
seg_list = jieba.cut('这里是天安门广场',cut_all=False)
```

```python
data_all=[]
temp={}
seg_list = []
for i in data:
    if i == '\n':
        data_all.append(temp)
        temp={}
    i=i.strip('\n')
    
    if i[:9] == 'Content: ':
        i=i.strip('Content: ')
        temp['Content: ']=i
        seg_list.append(list(jieba.cut_for_search(i)))
```
